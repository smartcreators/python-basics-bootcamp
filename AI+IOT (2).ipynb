```python
#step-3: handling the missing data or values.
#scenario-1: missing values in numeriacl colomn.
#calculate the mean or avarage of that colomn, and replace the empty values.
#scenario-2: missing values is non numerical colomns., which can be replaced with the highly repeated values.
#scenario-3:missing  values in huge data set.


import numpy as np
from sklearn.impute  import SimpleImputer  # it is a class which is uesd to replace missing values from sklearn .
imputer=SimpleImputer(missing_values=np.nan,strategy='mean') # imputer variable is created, where the nan value is replaced with the mean value and it is stored in imputer.
imputer.fit(x[:,1:3]) # for which colomns we want to apply the methods
x[:,1:3]=imputer.transform(x[:,1:3]) # the nan values are replaced with mean values
```

```python
x
```

```python
# step-4: handling the catagorical data.
# which converts the text data into numarical data
from sklearn.preprocessing import OneHotEncoder # OneHotEncoder is class which was imported from sklearn.preprocessing. it is converts text code into binary code. OneHotEncoder is only used for input column.
from sklearn.compose import ColumnTransformer #ColomnTransfermer is a class which was imported from sklearn.compose.  ColomnTransefermer is a class is used to apply the changes of catagorical data into binary data

#Alexa-0,0,1

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')   #encoder is a method which converts text to binary code.

x=ct.fit_transform(x)
```

```python
from sklearn.preprocessing import LabelEncoder # import the label encoder library from sklear.preprocessing
le=LabelEncoder() # the  LabelEncoder is used to encode the output columns.
y=le.fit_transform(y)


```

```python
#step-5: split the data into training and testing the data
#we have to import the library from sklearn and selection
from sklearn.model_selection import train_test_split  # train_test_split is a class which was imported  from sklearn.model_selection. it was used for splinting the data training and testing.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```
